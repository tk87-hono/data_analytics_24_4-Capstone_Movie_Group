```python
import cpi
import time
import numpy as np

def inflate_column(data, value_column, year_column, target_year=2023):
    """
    Adjust the values in the `value_column` of the dataframe `data` for inflation, using the `cpi` library.
    Inflates the values to the target year (default: 2023).
    """
    print(f'Beginning to inflate {value_column}')
    start_time = time.time()

    # Create arrays for the value and year columns
    values = data[value_column].values
    years = data[year_column].values

    # Apply the inflation using a list comprehension
    inflated_values = np.array([cpi.inflate(v, y, to=target_year) for v, y in zip(values, years)])
    
    # Assign the result back to the dataframe
    data['inflated_value'] = inflated_values
    
    print("Inflation process took", time.time() - start_time, "seconds to run")
    return data

# Call the function (assuming bx_df is your dataframe)
test_df_larger = inflate_column(test_df_larger, 'value', 'year')
```

```python
# Beispiel-DataFrame mit den Spalten "value" und "year"
data = {
    'value': [100, 200, 300],  # Werte in der Spalte "value"
    'year': [2010, 2015, 2020] # Jahre in der Spalte "year"
}
test_df = pd.DataFrame(data)

# 1. Einmalige Berechnung der Inflationsfaktoren für jedes Jahr
unique_years = test_df['year'].unique()
inflation_factors = {year: cpi.inflate(1, year, to=2023) for year in unique_years}

# 2. Zuordnung der Inflationsfaktoren zu den entsprechenden Jahren
test_df['inflation_factor'] = test_df['year'].map(inflation_factors)

# 3. Berechnung der neuen Spalte, die die inflatierten Werte enthält
test_df['newColumn'] = test_df['value'] * test_df['inflation_factor']

# Ausgabe des DataFrames mit der neuen Spalte
print(test_df)
```
