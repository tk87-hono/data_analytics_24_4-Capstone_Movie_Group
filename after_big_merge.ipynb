```python
# Correction of dtypes in imdb_df
# TO INTEGER

imdb_df[["runtime","num_votes", "genres_count", "directors_count", "writers_count", "EU_since_1996", "EU_tickets_sold", "NA_tickets_sold"]] = imdb_df[["runtime","num_votes", "genres_count", "directors_count", "writers_count", "EU_since_1996", "EU_tickets_sold", "NA_tickets_sold"]].astype("Int64")
imdb_df.info()

# TO BOOLEAN
imdb_df["is_adult"] = imdb_df["is_adult"].astype(bool)

## As there are just false values in the column we drop it
imdb_df.drop(columns="is_adult", inplace = True)
```

```python
display(bx_df)
display(bx_df.info())
```

RESULT (needed dtype corrections):
- [value] to integer
    - Stringstransf: replace "$" with ""
    - Stringstransf: replace "," with ""
- [release_group] to integer

```python
# Transformation of colum [value]
bx_df.loc[ : , "value"] = bx_df["value"].str.replace("$", "")
bx_df.loc[ : , "value"] = bx_df["value"].str.replace(",", "")
bx_df["value"] = bx_df["value"].astype("Int64")

# Transformation of colum [release_group]
bx_df["release_group"] = bx_df["release_group"].astype("Int64")
```

### Drop all rows that dont have sales in column [value]

```python
bx_df.dropna(inplace= True)
```

```python
bx_df.drop(bx_df[((bx_df["release_group"] == 0) | (bx_df["release_group"] == 1)) & (bx_df["region"] == "Domestic")].index, inplace = True)
```

```python
bx_df[(bx_df["release_group"] == 0) | (bx_df["release_group"] == 1)].groupby(["tconst", "region"]).sum()
```

```python
bx_df_new = bx_df[(bx_df["release_group"] == 0) | (bx_df["release_group"] == 1)].groupby(["tconst", "region"]).sum()
bx_df_new["release_group"] = 2
bx_df_new.reset_index(inplace=True)
```

```python
bx_df = pd.concat([bx_df,bx_df_new])
bx_df.drop(bx_df[(bx_df["release_group"] == 0) | (bx_df["release_group"] == 1)].index, inplace = True)
bx_df.drop(columns="release_group", inplace = True)
```

# Inflation Correction

Approach:
- 1) Use Boxoffice-table
- 2) Boxoffice-table: Re-Releases 
- 3) Add sales data AND year from imdb_df to Boxoffice-table
- 4) Use inflation-correction function on Sales-data on all columns (make new columns)
- 5) Drop Salescolums in imdb_df
- 6) Merge sales data back from Boxoffice-table to imdb_data

```python
# Merge sales AND year data to Boxoffice-table
bx_df = bx_df.merge(imdb_df[["tconst", "year", "NA_gross_sales"]], how = "inner", left_on = "tconst", right_on = "tconst")

# Resort columns
bx_df = bx_df.reindex(["tconst","region", "release_group", "year", "value", "NA_gross_sales"], axis=1)
```
