```python
comp_df = imdb_tick_df.merge(bx_df, on="tconst", how="left")
comp_df
```

```python
comp_df = comp_df.merge(genres_df, on="tconst", how="left")
comp_df
```

```python
comp_df = comp_df[comp_df['country'] == "World"]
```

```python
comp_df.drop(columns=["tconst", 'values', 'country', 'NA_tickets_sold', 'NA_gross_sales', 'distributor', 'EU_tickets_sold', 'EU_since_1996', 'writer3_name', 'writer2_name', 'writer_name', 'writers_count', 'director3_name', 'director2_name', 'director_name', 'directors_count', 'acting5', 'acting4', 'acting3', 'acting2', 'acting', 'genre3', 'genre2', 'genre_x', 'genres_count', 'average_rating', 'num_votes', 'runtime', 'original_title', 'primary_title'], inplace = True)
```

```python
comp_df = comp_df[["year","genre_y","inflated_values"]]
comp_df.rename(columns={"genre_y":"genre"}, inplace=True)
comp_df = comp_df[comp_df['genre'].notna()]
comp_df = comp_df.sort_values(['year', 'inflated_values'], ascending=[True, False])
```

```python
comp_df = comp_df.groupby(["year", "genre"]).mean().sort_values(["year", "inflated_values"], ascending=[True,False]).reset_index()
```

```python
comp_df_sorted = comp_df.sort_values(['year', 'inflated_values'], ascending=[True, False])
top_5_per_year_genre = comp_df_sorted.groupby(['year', 'genre']).head(5).reset_index(drop=True)
```
