## Import data

```python
import pandas as pd
```

```python
df_title_basics = pd.read_csv("Data/title.principals/title.basics.csv")
```

```python
df_title_basics.head()
```

```python
df_title_basics["titleType"].unique()
```

```python
movie_mask = df_title_basics["titleType"].isin(['movie', 'tvMovie'])
year_mask = df_title_basics["startYear"] >= 1950

combined_mask = (movie_mask) & (year_mask)
```

```python
df_title_basics_filtered = df_title_basics[combined_mask]
df_title_basics_filtered.shape
```

```python
df_title_basics_filtered.describe(include="all")
```

```python
df_title_basics_filtered["tconst"]
```

```python
df_title_basics_filtered["tconst"].to_csv("Data/tconst.csv", index= False)
```
